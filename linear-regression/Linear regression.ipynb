# Linear regression in plain Python

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

## The dataset that will be used in this example

```python
x = [5, 4, 3, 6, 8, 1, 9, 2, 7]
y = [10, 9, 6, 14, 14, 2, 17, 5, 13]
```

## Visualizing our dataset

```python
plt.plot(x,y, 'bo', label='x')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear dependency')
```

## Compute the required variables

```python
xy_sum = np.sum(np.multiply(x,y))
x_sum = np.sum(x)
y_sum = np.sum(y)
x2_sum = np.sum(np.power(x,2))
n = len(x)
```

## Compute the slope
$b = \frac{n\sum{xy} - \sum{x}\sum{y}}{n\sum{x^2} - (\sum{x})^2}$

```python
b_numerator = n * xy_sum - x_sum * y_sum
b_denominator = n * x2_sum - np.power(x_sum,2)
b = b_numerator / b_denominator
```

## Compute the coefficent
$a = \frac{\sum{y}-b\sum{x}}{n}$

```python
a = (y_sum - (b*x_sum)) / n
```

## Create a formula to compute the regression line
$y' = a+bx$

```python
regression = lambda x : a + b*x
```

# Visualize the results

```python
plt.plot(x, y, 'bo')
plt.plot(x, [regression(i) for i in x], 'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Linear dependency with trend line')
```
